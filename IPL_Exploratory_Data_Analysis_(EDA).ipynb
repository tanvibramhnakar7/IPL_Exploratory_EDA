# **IPL Exploratory Data Analysis (EDA)  Project**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches_df = pd.read_csv('/content/IPL Matches 2008-2020.csv')
balls_df = pd.read_csv('/content/IPL Ball-by-Ball 2008-2020.csv')
```

```python
matches_df['date'] = pd.to_datetime(matches_df['date'])
matches_df['season'] = matches_df['date'].dt.year
```

```python
matches_df.head(25)
```

```python
balls_df.head(20)
```

```python
# 1. What was the count of matches played in each season?
Matches_played_each_season = matches_df['season'].value_counts()
Matches_played_each_season
```

```python
# setting for the seaborn plot
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
plt.figure()
sns.barplot(x=Matches_played_each_season.index,y=Matches_played_each_season.values,palette="viridis",edgecolor="yellow")
plt.title('Matches played in each season')
plt.xlabel('Season')
plt.ylabel('Matches')
plt.xticks(rotation=35)
plt.tight_layout()
plt.show()
```

```python
# 2. How many runs were scored in each season?

merge_df = pd.merge(matches_df, balls_df, on='id')
runs_per_season = merge_df.groupby('season')['total_runs'].sum()
runs_per_season
```

```python
sns.barplot(x=runs_per_season.index,y=runs_per_season.values,palette="plasma",edgecolor="black")
plt.title('runs were scored in each season')
plt.xlabel('Season')
plt.ylabel('Total runs')
plt.xticks(rotation=25)
plt.show()
```

```python
# 3. What were the runs scored per match in different seasons?
merge_df = pd.merge(matches_df, balls_df, on='id')
runs_per_season = merge_df.groupby(['season', 'id'])['total_runs'].sum()
runs_per_season

# Calculate the average runs per match for each season
avg_runs_per_match = runs_per_season.groupby('season').mean()
avg_runs_per_match
```

```python
sns.barplot(x=avg_runs_per_match.index, y=avg_runs_per_match.values, palette="viridis",edgecolor="black")
plt.title('Average runs scored per match in each season')
plt.xlabel('Season')
plt.ylabel('Average Total Runs')
plt.xticks(rotation=25)
plt.show()
```

```python
# 4. Who has umpired the most?

most_matches_umpired =(matches_df['umpire1'].value_counts() + matches_df['umpire2'].value_counts()).idxmax()
print(f'{most_matches_umpired} had umpired the most matches in ipl.')
```

```python
# 5. Which team has won the most tosses?
most_toss_win=(matches_df['toss_winner'].value_counts()).idxmax()
print(f'{most_toss_win} had won the most tosses.')
```

```python
# 6. What does the team decide after winning the toss?
decision = matches_df['toss_decision'].value_counts()
decision
```

```python
plt.figure(figsize=(9,7))
plt.pie(decision, labels=decision.index, autopct='%1.1f%%', startangle=45)
plt.title('Toss Decision Distribution')
plt.show()
```

```python
# 7. How does the toss decision vary across seasons?
toss_decision = matches_df.groupby(['season','toss_decision'])['id'].count().unstack()
toss_decision
```

```python
toss_decision.plot(kind='bar',stacked=True,colormap="Set3")
plt.title('Toss Decision Across Seasons')
plt.xlabel('Season')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 8. Does winning the toss imply winning the game?
matches_df["Toss_win_match"] = matches_df['toss_winner'] == matches_df['winner']
win_rate = matches_df["Toss_win_match"].mean()
win_rate

Toss_Outcome = matches_df["Toss_win_match"].value_counts()
Toss_Outcome = Toss_Outcome.rename({False: "Loss", True: "Win"})
Toss_Outcome
```

```python
plt.figure(figsize=(5,5))
Toss_Outcome.index=['Toss Won and Match Won','Toss Won and Match Lost']
Toss_Outcome.plot(kind="pie", autopct="%.3f%%", startangle=50, colors=["green","red"])
plt.title("Toss Won and Match Won")
plt.show()
```

```python
# 9. How many times has the chasing team won the match?
chasing_team_won=matches_df[(matches_df['toss_decision'] == 'field')& (matches_df['toss_winner']==matches_df['winner'])].shape[0]
print(f"The chasing team has won the match {chasing_team_won} times.")
```

```python
# 10. Which all teams had won this tournament?
final_match=matches_df.sort_values('date').drop_duplicates('season',keep="last")
tournament_winner=final_match['winner'].value_counts()
tournament_winner
```

```python
plt.figure(figsize=(5,5))
tournament_winner.plot(kind="pie", autopct="%.2f%%", startangle=50, colors=["green","red",'blue','yellow','skyblue','brown'])
plt.title("Teams Won the Tournament")
plt.show()
```

```python
# 11. Which team has played the most number of matches
most_matches_played = (matches_df['team1'].value_counts() + matches_df['team2'].value_counts()).idxmax()
print(f'Across all seasons most matches played by {most_matches_played}.')
```

```python
# 12. Which team has won the most number of times?
most_win=matches_df['winner'].value_counts().idxmax()
print(f'Most number of times team Won is {most_win}.')
```

```python
# 13. Which team has the highest winning percentage?
total_matches=matches_df['team1'].value_counts().sort_index()+ matches_df['team2'].value_counts().sort_index()
wins=matches_df['winner'].value_counts().sort_index()
winning_percentage=(wins / total_matches) * 100
higgest_winning_percentage= winning_percentage.idxmax()
print(f'The team with the highest winning percentage is: {higgest_winning_percentage}')
```

```python
# 14. Is there any lucky venue for a particular team?
lucky_venue = matches_df.groupby(['venue', 'winner']).size().reset_index(name='wins')
lucky_venue_sorted = lucky_venue.sort_values('wins', ascending=False)
lucky_venue_for_teams = lucky_venue_sorted.groupby('winner').first().sort_values(by='wins',ascending=False)
display(lucky_venue_for_teams)
```

```python
# 15. Innings wise comparison between teams
inning_comparison = balls_df.groupby(['batting_team','inning'])['total_runs'].sum().unstack()
inning_comparison
```

```python
# 16. Which team has scored the most number of 200+ scores?
most_score = (balls_df.groupby(["id","batting_team"])["total_runs"].sum()
            .reset_index().query("total_runs >= 200").groupby("batting_team").size().sort_values(ascending=False)).idxmax()
print(f'{most_score}, has scored most number of 200.')
```

```python
# 17. Which team has conceded 200+ scores the most?
most_conceded = (balls_df.groupby(["id","bowling_team"])["total_runs"].sum().reset_index()
            .query("total_runs >= 200").groupby("bowling_team").size().sort_values(ascending=False)).idxmax()
print(f'{most_conceded}, has conceded most number of 200.')
```

```python
# 18. What was the highest run scored by a team in a single match?

df_scores = balls_df.groupby(['id', 'batting_team'])['total_runs'].sum().reset_index()
highest_score= df_scores.loc[df_scores['total_runs'].idxmax()]
highest_score
```

```python
# 19. Which is the biggest win in terms of run margin?
matches_df['result_margin']=pd.to_numeric(matches_df['result_margin'], errors='coerce')
biggest_win=matches_df[['winner','team1','team2','result_margin']][matches_df['result_margin']==matches_df['result_margin'].max()]
biggest_win
```

```python
# 20. Which batsmen have played the most number of balls?
bat = balls_df.groupby("batsman")["ball"].count().sort_values(ascending=False).idxmax()
print(f'The most ball played by {bat}.')
```

```python
# 21. Who are the leading run-scorers of all time?
bat = balls_df.groupby("batsman")["batsman_runs"].sum().sort_values (ascending=False).idxmax()
print(f"The leading run-scorers of all time is {bat}")
```

```python
# 22. Who has hit the most number of 4's?
most_fours=balls_df[balls_df["batsman_runs"] == 4]["batsman"].value_counts().idxmax()
print(f'Most fours hit by {most_fours}')
```

```python
# 23. Who has hit the most number of 6's?
most_sixes=balls_df[balls_df["batsman_runs"] == 6]["batsman"].value_counts().idxmax()
print(f'Most sixes hit by {most_sixes}')
```

```python
# 24. Who has the highest strike rate?
strike_rate = balls_df.groupby('batsman')['batsman_runs'].sum()
ball = balls_df.groupby('batsman')['ball'].count()
strike_rate = strike_rate / ball * 100
higgest_strike_rate=strike_rate.idxmax()
print(f'{higgest_strike_rate}, has the highest strike rate.')
```

```python
# 25. Who is the leading wicket-taker?
leading_wicket_taker = balls_df.groupby("bowler")["dismissal_kind"].count().sort_values(ascending=False).idxmax()
print(f'The leading wicket-taker of all time is {leading_wicket_taker}')
```

```python
# 26. Which stadium has hosted the most number of matches?
stadium_hosted_most_matches=matches_df['venue'].value_counts().idxmax()
print(f'{stadium_hosted_most_matches}, has hosted most number of matches.')
```

```python
# 27. Who has won the most MOM awards?
MOM= matches_df['player_of_match'].value_counts().idxmax()
print(f"{MOM}, has won the most MOM awards.")
```

```python
# 28. What is the count of fours hit in each season?
merge_df = pd.merge(matches_df, balls_df, on='id')

fours_per_season = merge_df[merge_df['batsman_runs'] == 4].groupby('season')['batsman_runs'].count()
fours_per_season
```

```python
plt.figure(figsize=(9,5))
sns.barplot(x=fours_per_season.index, y = fours_per_season.values, palette="viridis",edgecolor="black")
plt.xlabel('Season')
plt.ylabel('Number of Fours')
plt.title('Count of Fours Hit in Each Season')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 29. What is the count of sixes hit in each season?
merge_df = pd.merge(matches_df, balls_df, on='id')

sixes_per_season = merge_df[merge_df['batsman_runs'] == 6].groupby('season')['batsman_runs'].count()
sixes_per_season
```

```python
plt.figure(figsize=(9,5))
sns.barplot(x=sixes_per_season.index, y = sixes_per_season.values, palette="viridis",edgecolor="black")
plt.xlabel('Season')
plt.ylabel('Number of Sixes')
plt.title('Count of Sixes Hit in Each Season')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 30. What is the count of runs scored from boundaries in each season

fours_per_season = merge_df[merge_df['batsman_runs'] == 4].groupby('season')['batsman_runs'].count()
sixes_per_season = merge_df[merge_df['batsman_runs'] == 6].groupby('season')['batsman_runs'].count()
boundaries = (fours_per_season * 4 ) + (sixes_per_season * 6)

boundaries
```

```python
plt.figure(figsize=(9,5))
sns.barplot(x=boundaries.index, y =boundaries.values, palette="viridis",edgecolor="black")
plt.xlabel('Season')
plt.ylabel('Number of Runs')
plt.title('Count of Runs Scored from Boundaries in Each Season')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 31. What is the run contribution from boundaries in each season?
contribution = merge_df.groupby('season')['total_runs'].sum()
contribution = (boundaries / contribution) * 100
contribution
```

```python
# 32. Which team has scored the most runs in the first 6 overs?
first_six_over = balls_df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False).idxmax()
print(f'{first_six_over}, has scored most runs in first six overs.')
```

```python
# 33. Which team has scored the most runs in the last 4 overs?

last_4_over=balls_df[balls_df['over']>=17]
total_run_in_last_4_over=last_4_over.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False).idxmax()
print(f"{total_run_in_last_4_over}, has scored most runs in last four overs.")
```

```python
# 34. Which team has the best scoring run-rate in the first 6 overs?

pp_run_rate=(balls_df[balls_df['over']<7].groupby('batting_team')['total_runs'].sum() /6).idxmax()
print(f"{pp_run_rate}, has the best scoring run-rate in the first six overs")
```

```python
# 35. Which team has the best scoring run-rate in the last 4 overs?
last_four_overs_run_rate =(balls_df[balls_df['over']>15].groupby('batting_team')['total_runs'].sum()/4).idxmax()
print(f'{last_four_overs_run_rate}, has the best run-rate in the last four overs.')
```
